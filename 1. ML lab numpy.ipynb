**data types in  NumPy**


* i - integer
* b - boolean
* u - unsigned int
* f - float
* c - complex float
* m - time delta
* M - datetime
* O - object
* S - string

#COPYING ARRAYS

```python
a = np.array([1,2,3])
b=a #a is created and then b is also pointing to same object, a copy is not getting created its the same object
print(b)
b
b[0]=100
print(b)
print(a)
```

```python
b = a.copy() #here the content of a is copied in b
print(a)
b[2]=90
print(b)
print(a)
```
