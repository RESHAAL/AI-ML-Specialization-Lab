**Multi-variate EDA**

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('/content/drive/My Drive/New/train.csv')
```

```python
titanic.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue = 'sex', style = 'smoker', size = 'size', data = df)
```

```python
sns.barplot(x = 'Pclass', y = 'Age', hue = 'Sex', data = titanic)
```

```python
sns.boxplot(x = 'Pclass', y = 'Age', hue = 'Sex', data = titanic)
```

```python
sns.displot(titanic['Age'])
```

```python
sns.displot(titanic[titanic['Survived'] == 0]['Age'])
sns.displot(titanic[titanic['Survived'] == 1]['Age'])
```

```python
(titanic.groupby('Pclass')['Survived'].mean())*100
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```
